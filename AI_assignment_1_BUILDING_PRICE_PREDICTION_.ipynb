```python
#multivariate analysis
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
data = pd.read_csv("/content/drive/My Drive/House Price India.csv")
data = data.dropna()
corr_matrix = data.corr()
print(corr_matrix)
corr_coef, p_value = pearsonr(data['Built Year'], data['Price'])
print("Correlation coefficient:", corr_coef)
print("P-value:", p_value)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)
print(pca.explained_variance_ratio_)
import matplotlib.pyplot as plt

plt.scatter(principal_components[:,0], principal_components[:,1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
#task4 descriptive statistics
import pandas as pd
csv_file_path = '/content/drive/My Drive/House Price India.csv'
chunk_size = 10000
count = 0
mean = 0
M2 = 0
for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    x = chunk['Price']
    n = len(x)
    delta = x - mean
    mean += delta.sum() / n
    M2 += (delta * (x - mean)).sum()
    count += n
variance = M2 / (count - 1)
stddev = variance ** 0.5
mean = mean
print(f"Mean: {mean:.2f}")
print(f"Standard deviation: {stddev:.2f}")
```
